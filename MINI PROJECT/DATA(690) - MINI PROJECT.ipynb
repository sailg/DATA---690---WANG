 I am going to check that Horse power of a car and its mileage are directly proportional to each other.

```python
list1 = df["Engine Information.Engine Statistics.Horsepower"]/df["Fuel Information.City mpg"] 
```

```python
import matplotlib.pyplot as plt
# Adjusting size of the graph
plt.figure(figsize=(10,5))
# Line plot of list1 
plt.plot(list1)
plt.show()
```

```python
# Histogram of list1 values
plt.hist(list1, 20,density=True,histtype='bar',alpha=0.5)
plt.show()
```

From the above plot, we can see that almost 70% of the list1 values lies in between the range of 5-30. And the varience is not too high , so we can assume that as the torque value increases the milage of the car is also getting increases.

```python
df['Fuel Information.Highway mpg'].describe()
```

```python
list2 = df["Fuel Information.Highway mpg"]-df["Fuel Information.City mpg"]
```

```python
# Adding the list2 as a new column in the dataframe.
df['Difference between mpg in highway and city'] = np.array(list2)
```

```python
df.head()
```

```python
# Getting the unique values from 'Engine Information.Number of Forward Gears' column.
x=(df['Engine Information.Number of Forward Gears'].unique())
# Plotting the mean of values from 'Difference between mpg in highway and city' column group by 'Engine Information.Number of Forward Gears' column.
plt.ylabel("Mean of difference b/w mpg in highway and city")
(df.groupby(['Engine Information.Number of Forward Gears'])['Difference between mpg in highway and city'].mean()).plot.bar()
```

From the above bar graph, we can judge that as the number of gears increases the difference between milag per gallon in highway and city increases except in case from gear 6 to 7.

```python
# Scatter plot of Horsepower (vs) Torque
df.plot.scatter(x ='Engine Information.Engine Statistics.Horsepower', y ='Engine Information.Engine Statistics.Torque') 
plt.show()
```

From this scatter plot we can sence that Horsepower and Torque have moderate linear relationship.

Mean of milage per gallon of all cars grouped by year.

```python
# Bar plot of year (vs) Mean of values in 'Fuel Information.Highway mpg' column group by year.
plt.ylabel("Mean of mpg in highways")
(df.groupby(['Identification.Year'])['Fuel Information.Highway mpg'].mean()).plot.bar()
```

Cars (vs)  Its Average Mileage

```python
# Taking the unique values of Identification.Make column into list x.
x= df['Identification.Make'].unique()
# Mean values of highay mileage group by Identification.Make.
y= df.groupby(['Identification.Make'])['Fuel Information.Highway mpg'].mean()
# Defining size of graph
f, ax = plt.subplots(figsize=(50,10))
plt.xticks(rotation='vertical',size=30)
plt.yticks(size=30)
# Labeling the y-axis
plt.ylabel("Average Mileage",size=30)
# Labeling the x-axis
plt.xlabel("Name of the car",size=30)
# Plotting the bar graph with bar colors as green
plt.bar(x,y,color='green')
plt.show()
```
